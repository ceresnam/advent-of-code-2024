````python
from typing import Literal, TypeAlias

Cell = Literal['.', '#', 'O']
Grid: TypeAlias = list[list[Cell]]
Position = tuple[int, int]
Path = list[Position]

def draw_path_in_grid(grid: Grid, path: Path) -> str:
    grid = [row.copy() for row in grid]

    prev_pos = None
    for row, col in path:
        if grid[row][col] in {'^','#','O'}:
            continue

        if prev_pos is None:
            grid[row][col] = '^'
        elif row == prev_pos[0] and col == prev_pos[1]: # turn
            grid[row][col] = '+'  
        elif row == prev_pos[0]: # horizontal
            grid[row][col] = '-'
        elif col == prev_pos[1]: # vertical
            grid[row][col] = '|'
        else:
            grid[row][col] = '?'
        prev_pos = (row, col)      
    return '\n'.join(''.join(row) for row in grid)
````

````python
def is_loop(grid: Grid, start_pos: Position):
    rows, cols = len(grid), len(grid[0])
    max_steps = rows * cols * 4  # 4 is the number of directions
    directions = [(-1, 0), (0, 1), (1, 0), (0, -1)]    
    visited_states = set()  # (pos, direction) pairs
    current_pos = start_pos
    current_dir_idx = 0  # up
    steps = 0
    
    while steps < max_steps:
        state = (current_pos, current_dir_idx)
        if state in visited_states:
            return True
        
        visited_states.add(state)
        row, col = current_pos
        next_row = row + directions[current_dir_idx][0]
        next_col = col + directions[current_dir_idx][1]
        
        # Check if out of bounds
        if (next_row < 0 or next_row >= rows or 
            next_col < 0 or next_col >= cols):
            return False
            
        # If obstacle ahead, turn right
        if grid[next_row][next_col] == '#' or grid[next_row][next_col] == 'O':
            current_dir_idx = (current_dir_idx + 1) % 4
        else:
            current_pos = (next_row, next_col)
        
        steps += 1
    
    return False  # If we reach max_steps, assume it's a loop

def find_loop_positions(grid_str):
    # Convert string to 2D grid
    grid = [list(line) for line in grid_str.strip().split('\n')]
    rows, cols = len(grid), len(grid[0])
    print(rows, cols, rows*cols)
    
    # Find starting position
    start_pos = None
    for i in range(rows):
        for j in range(cols):
            if grid[i][j] == '^':
                start_pos = (i, j)
                break
        if start_pos:
            break
    

    loop_positions = []
    
    # Try each empty position
    for i in range(rows):
        for j in range(cols):
            if grid[i][j] == '.' and (i, j) != start_pos:
                # Try placing obstacle
                grid[i][j] = 'O'
                if is_loop(grid, start_pos):
                    loop_positions.append((i, j))
                grid[i][j] = '.'
    
    return len(loop_positions)

def test_part2():
    test_input = """....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#..."""

    result = find_loop_positions(test_input)
    assert result == 6, f"Expected 6, got {result}"
    print("Test passed!")

test_part2()

# Solve part 2
result = find_loop_positions(full_input)
print(f"Part 2 solution: {result}")
````
