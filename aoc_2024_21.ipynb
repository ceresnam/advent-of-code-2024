def get_typing_sequences(G: nx.Graph, code: str, start_key='A') -> list:
    if not code:
        return []
    
    result = []
    curr_pos = next(n for n, d in G.nodes(data=True) if d['key'] == start_key)
    
    for target_key in code:
        # Get all paths to the target key
        paths = get_paths(G, G.nodes[curr_pos]['key'], target_key)
