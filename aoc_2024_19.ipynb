def build_graph(patterns, target):
    """
    Build a directed graph where:
    - nodes are positions in the target string
    - edges represent valid pattern matches
    """
    G = nx.DiGraph()
    
    # Add nodes for each position (0 to len(target))
    for i in range(len(target) + 1):
        G.add_node(i)
    
    # For each starting position
    for start in range(len(target)):
        # Try each pattern
        for pattern in patterns:
            end = start + len(pattern)
            # Check if pattern fits within target
            if end <= len(target):
                # Check if pattern matches at this position
                if target[start:end] == pattern:
                    G.add_edge(start, end, pattern=pattern)
    
    return G

````python
import networkx as nx

# Test data from problem description
test_patterns = ['r', 'wr', 'b', 'g', 'bwu', 'rb', 'gb', 'br']
test_designs = [
    'brwrr',    # possible
    'bggr',     # possible
    'gbbr',     # possible
    'rrbgbr',   # possible
    'ubwu',     # impossible
    'bwurrg',   # possible
    'brgr',     # possible
    'bbrgwb'    # impossible
]

def build_graph(patterns, target):
    """
    Build a directed graph where:
    - nodes are positions in the target string
    - edges represent valid pattern matches
    """
    G = nx.DiGraph()
    
    # Add nodes for each position (0 to len(target))
    for i in range(len(target) + 1):
        G.add_node(i)
    
    # For each starting position
    for start in range(len(target)):
        # Try each pattern
        for pattern in patterns:
            end = start + len(pattern)
            # Check if pattern fits within target
            if end <= len(target):
                # Check if pattern matches at this position
                if target[start:end] == pattern:
                    G.add_edge(start, end, pattern=pattern)
    
    return G

def can_make_design(patterns, design):
    """
    Check if a design can be made using the available patterns
    Returns True if design can be made, False otherwise
    """
    G = build_graph(patterns, design)
    # Check if there's a path from start (0) to end (len(design))
    try:
        path = nx.shortest_path(G, 0, len(design))
        return True
    except nx.NetworkXNoPath:
        return False

# Test all designs
for i, design in enumerate(test_designs):
    result = can_make_design(test_patterns, design)
    print(f"Design {i+1}: '{design}' -> {'possible' if result else 'impossible'}")
    
# Count total possible designs
total_possible = sum(1 for d in test_designs if can_make_design(test_patterns, d))
print(f"\nTotal possible designs: {total_possible}")
````

import numpy as np

def count_ways_matrix(patterns, design):
    """
    Count ways using matrix approach
    """
    G = build_graph(patterns, design)
    n = len(design) + 1
    
    # Create adjacency matrix
    A = np.zeros((n, n), dtype=int)
    for u, v in G.edges():
        A[u][v] = 1
    
    # Initialize result array for paths of different lengths
    paths = np.zeros(n, dtype=int)
    
    # Current state matrix
    current = A.copy()
    paths[1] = current[0][n-1]  # paths of length 1
    
    # Count paths of each length
    for length in range(2, n):
        current = current @ A  # matrix multiplication
        paths[length] = current[0][n-1]
    
    return sum(paths)

````python
import numpy as np

def count_ways_matrix(patterns, design):
    """
    Count ways using matrix approach
    """
    G = build_graph(patterns, design)
    n = len(design) + 1
    
    # Create adjacency matrix
    A = np.zeros((n, n), dtype=int)
    for u, v in G.edges():
        A[u][v] = 1
    
    # Initialize result array for paths of different lengths
    paths = np.zeros(n, dtype=int)
    
    # Current state matrix
    current = A.copy()
    paths[1] = current[0][n-1]  # paths of length 1
    
    # Count paths of each length
    for length in range(2, n):
        current = current @ A  # matrix multiplication
        paths[length] = current[0][n-1]
    
    return sum(paths)

# Test with examples first
for i, design in enumerate(test_designs):
    ways = count_ways_matrix(test_patterns, design)
    print(f"Design {i+1}: '{design}' -> {ways} ways")

print(f"\nTotal ways: {sum(count_ways_matrix(test_patterns, d) for d in test_designs)}")
````
