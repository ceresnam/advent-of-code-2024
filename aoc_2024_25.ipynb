````python
def parse_input(input_text):
    locks = []
    keys = []
    dimensions = set()
    
    # Split into individual schematics
    schematics = input_text.split('\n\n')
    
    for schematic in schematics:
        lines = schematic.splitlines()
        heights = []
        width = len(lines[0])
        dimensions.add((width, len(lines)))
        
        # Calculate heights for each column
        for col in range(width):
            # Count from top until we find a '.'
            if lines[0][col] == '#':  # It's a lock
                height = 0
                for row in range(len(lines)):
                    if lines[row][col] == '.':
                        break
                    height += 1
                heights.append(height - 1)  # -1 because we count from 0
            else:  # It's a key
                height = 0
                for row in range(len(lines)-2, -1, -1):
                    if lines[row][col] == '#':
                        height += 1
                heights.append(height)
        
        # Determine if it's a lock or key and add to appropriate list
        if lines[0][0] == '#':
            locks.append(tuple(heights))
        else:
            keys.append(tuple(heights))
            
    return locks, keys, dimensions

# Test function
def test_parse_input():
    locks, keys, dimensions = parse_input(sample_input)
    print("Locks:", locks)
    print("Keys:", keys)
    print("Dimensions:", dimensions)
    # Verify against example values from problem description
    assert locks == [(0,5,3,4,3), (1,2,0,5,3)], "Locks don't match example"
    assert keys == [(5,0,2,1,3), (4,3,4,0,2), (3,0,2,0,1)], "Keys don't match example"
    print("All tests passed!")

test_parse_input()
````

````python
def fit_together(lock, key):
    # For each column, check if lock + key heights fit within space
    for lock_height, key_height in zip(lock, key):
        if lock_height + key_height + 2 > 7:
            return False
    return True

# Let's test with the example data
def test_fit_together():
    # Test cases from problem description
    assert fit_together((0,5,3,4,3), (5,0,2,1,3)) == False  # "overlap in the last column"
    assert fit_together((0,5,3,4,3), (4,3,4,0,2)) == False  # "overlap in the second column"
    assert fit_together((0,5,3,4,3), (3,0,2,0,1)) == True   # "all columns fit"
    print("All tests passed!")

test_fit_together()
````
